```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df = pd.read_csv('Assi_3_sales_data_with_discounts.csv')



# Numerical columns
num_cols = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
```

```python
# 1. Descriptive Statistics
print("DESCRIPTIVE STATISTICS:")
print(df[num_cols].describe().round(2))
```

```python

# Histograms
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

for i, col in enumerate(num_cols):
    axes[i//3, i%3].hist(df[col], bins=25, edgecolor='black', alpha=0.7)
    axes[i//3, i%3].set_title(col)

plt.suptitle("Histograms", fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 3. Boxplots
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
for i, col in enumerate(num_cols):
    axes[i//3, i%3].boxplot(df[col])
    axes[i//3, i%3].set_title(col)
plt.suptitle("Boxplots", fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 4. Bar Charts for Categorical
cat_cols = ['Day', 'BU', 'Brand']
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for i, col in enumerate(cat_cols):
    df[col].value_counts().plot(kind='bar', ax=axes[i], color='coral', edgecolor='black')
    axes[i].set_title(f'{col} Distribution')
    axes[i].tick_params(rotation=45)
plt.tight_layout()
plt.show()
```

```python


# Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[num_cols].corr(), annot=True, fmt='.2f', cmap='RdBu_r', center=0)
plt.title("Correlation Heatmap", fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Conclusion
print(f"Total Revenue: ₹{df['Net Sales Value'].sum():,.0f}")
print(f"Avg Discount: {df['Discount Rate (%)'].mean():.2f}%")
print(f"Busiest Day: {df['Day'].value_counts().index[0]}")
print(f"Top Brand: {df['Brand'].value_counts().index[0]}")
print(f"\nStrongest Correlation: Volume ↔ Sales Value ({df['Volume'].corr(df['Total Sales Value']):.2f})")
```
