```python
"""

BASIC STATISTICS-2
Problem Statement : Hospital Patient Data Analysis
Context:
A hospital maintains patient records including admission details, department, diagnosis, doctor, and bill amount. You have two datasets: one with patient info and another with billing details. Some patients have blank bill amounts, and there are multiple rows for the same patient due to follow-ups.
Tasks:
1.	Load the patient dataset and show summary with info().
2.	Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].
3.	Drop administrative columns like ['ReceptionistID', 'CheckInTime'].
4.	Use groupby to find total bill amount per department.
5.	Remove duplicate patient records based on PatientID.
6.	Fill missing BillAmount values with the mean bill amount.
7.	Merge the billing dataset with patient dataset on PatientID.
8.	Concatenate an additional DataFrame that contains new patients for the current week (row-wise).
9.	Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).
Expected Outcome:
•	Final cleaned dataset with accurate billing info.
•	All missing values handled, merged dataset across PatientID.
•	Ability to perform further analytics on department-wise revenue or doctor performance.


"""
```

```python
import pandas as pd
import numpy as np
```

```python
# Load datasets

patient_df = pd.read_csv('Assi_4_Patient_Data.csv')
billing_df = pd.read_csv('Assi_4_Billing_Data.csv')
```

```python
# Task 1: Load and show summary

print("\n1. Dataset Info:")
print(patient_df.info())
```

```python
# Task 2 & 3: Select relevant columns and drop administrative columns

patient_clean = patient_df[['PatientID', 'Department', 'Doctor', 'BillAmount']].copy()
print("\n2. Selected Columns (Patient data):")
print(patient_clean.head())
```

```python
# Task 4: Groupby to find total bill per department

dept_bill = patient_clean.groupby('Department')['BillAmount'].sum()
print("\n3. Total Bill Amount per Department:")
print(dept_bill)
```

```python
# Task 5: Remove duplicate patient records

patient_unique = patient_clean.drop_duplicates(subset=['PatientID'])
print(f"\n4. Duplicates removed: {len(patient_clean) - len(patient_unique)} rows")
```

```python
# Task 6: Fill missing BillAmount with mean

mean_bill = patient_unique['BillAmount'].mean()
patient_unique['BillAmount'] = patient_unique['BillAmount'].fillna(mean_bill)
print(f"\n5. Missing values filled with mean: {mean_bill:.2f}")
```

```python
# Task 7: Merge with billing dataset

merged_df = pd.merge(patient_unique, billing_df, on='PatientID', how='left')
print("\n6. Merged Dataset:")
print(merged_df)
```

```python
import pandas as pd

# Example merged_df (make sure this exists)
# Remove this if you already created merged_df earlier
merged_df = pd.DataFrame({
    'PatientID': [101, 102],
    'Department': ['Cardiology', 'Orthopedics'],
    'Doctor': ['Dr. A', 'Dr. B'],
    'BillAmount': [5000, 6000]
})

# New patients
new_patients = pd.DataFrame({
    'PatientID': [106, 107],
    'Department': ['Neurology', 'Cardiology'],
    'Doctor': ['Dr. John', 'Dr. Smith'],
    'BillAmount': [4500, 6800]
})

# Concatenate
final_df = pd.concat([merged_df, new_patients], ignore_index=True)

print("\n7. After adding new patients (row-wise):")
print(final_df)
```

```python
# Task 9: Concatenate new billing category columns (column-wise)
new_columns = pd.DataFrame({
    'InsuranceCovered': [2000, 1500, 2500, 3000, 1000, 1800, 2200],
    'FinalAmount': [3000, 3500, 5000, 3200, 4000, 5300, 6200]
})

final_df = pd.concat([final_df, new_columns], axis=1)
print("\n8. Final Dataset with new billing columns:")
print(final_df)
```

```python
# Final summary

print("FINAL SUMMARY")

print(f"Total Patients: {len(final_df)}")
print(f"Total Revenue: ₹{final_df['BillAmount'].sum():,.2f}")
print(f"Department-wise Revenue:\n{final_df.groupby('Department')['BillAmount'].sum()}")
```
