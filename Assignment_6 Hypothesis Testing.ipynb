```python
"""
# Hypothesis Testing

Background:
Bombay hospitality Ltd. operates a franchise model for producing exotic Norwegian dinners throughout New England. The operating cost for a franchise in a week (W) is given by the equation W = $1,000 + $5X, where X represents the number of units produced in a week. Recent feedback from restaurant owners suggests that this cost model may no longer be accurate, as their observed weekly operating costs are higher.
Objective:
To investigate the restaurant owners' claim about the increase in weekly operating costs using hypothesis testing.
Data Provided:
●	The theoretical weekly operating cost model: W = $1,000 + $5X
●	Sample of 25 restaurants with a mean weekly cost of Rs. 3,050
●	Number of units produced in a week (X) follows a normal distribution with a mean (μ) of 600 units and a standard deviation (σ) of 25 units
Assignment Tasks:
1. State the Hypotheses statement:
2. Calculate the Test Statistic:
Use the following formula to calculate the test statistic(z):
where:
●	ˉxˉ = sample mean weekly cost (Rs. 3,050)
●	μ = theoretical mean weekly cost according to the cost model (W = $1,000 + $5X for X = 600 units)
●	σ = 5*25 units
●	n = sample size (25 restaurants)
3. Determine the Probability and compare:
Using the alpha level of 5% (α = 0.05),
4. Make a Decision:
Compare the test statistic with the critical value to decide whether to reject the null hypothesis.
5. Conclusion:
Based on the decision in step 4, conclude whether there is strong evidence to support the restaurant owners' claim that the weekly operating costs are higher than the model suggests.

Submission Guidelines:
●	Prepare a python file detailing each step of your hypothesis testing process.
●	Include calculations for the test statistic and the critical value.
●	Provide a clear conclusion based on your analysis.

"""
```

```python
import numpy as np
import scipy.stats as stats
import math
```

```python
# Data: durability in millions of characters
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 
                 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python
# Sample statistics
n = len(data)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
```

```python
# For 99% confidence, alpha = 0.01, so alpha/2 = 0.005
confidence_level = 0.99
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
```

```python
# Margin of error
margin_error = t_critical * (sample_std / math.sqrt(n))

# Confidence interval
ci_lower = sample_mean - margin_error
ci_upper = sample_mean + margin_error
```

```python
print(f"\nCritical t-value (t_{alpha/2, {n-1}}): {t_critical:.4f}")
print(f"Margin of Error: {margin_error:.4f}")
print(f"\n99% Confidence Interval: ({ci_lower:.4f}, {ci_upper:.4f}) million characters")

print("\n--- Explanation ---")
print("The t-distribution is used instead of normal distribution because:")
print("1. Population standard deviation is unknown")
print("2. Sample size is small (n=15 < 30)")
print("3. t-distribution provides wider intervals to account for sampling variability")
```

```python
# PART A: 99% Confidence Interval (Sample Standard Deviation)
```

```python
"""
Sample Size (n): 15
Sample Mean (x̄): 1.2380 million characters
Sample Standard Deviation (s): 0.1932 million characters
Degrees of Freedom (df): 14

Critical t-value (t_0.005, 14): 2.9768
Margin of Error: 0.1485

99% Confidence Interval: (1.0895, 1.3865) million characters

--- Explanation ---
The t-distribution is used instead of normal distribution because:
1. Population standard deviation is unknown
2. Sample size is small (n=15 < 30)
3. t-distribution provides wider intervals to account for sampling variability
"""
```

```python
# Known population standard deviation
population_std = 0.2  # million characters

print(f"Sample Mean (x̄): {sample_mean:.4f} million characters")
print(f"Population Standard Deviation (σ): {population_std} million characters")
print(f"Sample Size (n): {n}")

# For 99% confidence, use z-distribution
z_critical = stats.norm.ppf(1 - alpha/2)

# Margin of error using z
margin_error_z = z_critical * (population_std / math.sqrt(n))

# Confidence interval
ci_lower_z = sample_mean - margin_error_z
ci_upper_z = sample_mean + margin_error_z

print(f"\nCritical z-value (z_{alpha/2}): {z_critical:.4f}")
print(f"Margin of Error: {margin_error_z:.4f}")
print(f"\n99% Confidence Interval: ({ci_lower_z:.4f}, {ci_upper_z:.4f}) million characters")

print("\n--- Comparison ---")
print(f"With sample std (t-distribution): ({ci_lower:.4f}, {ci_upper:.4f})")
print(f"With population std (z-distribution): ({ci_lower_z:.4f}, {ci_upper_z:.4f})")
print(f"Interval width (t): {ci_upper - ci_lower:.4f}")
print(f"Interval width (z): {ci_upper_z - ci_lower_z:.4f}")
print("The t-distribution gives a wider interval due to uncertainty about population standard deviation")
```

```python
# PART B: 99% Confidence Interval (Known Population Standard Deviation)
```

```python
"""
Sample Mean (x̄): 1.2380 million characters
Population Standard Deviation (σ): 0.2 million characters
Sample Size (n): 15

Critical z-value (z_0.005): 2.5758
Margin of Error: 0.1330

99% Confidence Interval: (1.1050, 1.3710) million characters

--- Comparison ---
With sample std (t-distribution): (1.0895, 1.3865)
With population std (z-distribution): (1.1050, 1.3710)
Interval width (t): 0.2970
Interval width (z): 0.2660
The t-distribution gives a wider interval due to uncertainty about population standard deviation

"""
```

```python
# Given data
sample_mean_cost = 3050 
sample_size = 25
population_std_X = 25  
mean_X = 600  
alpha_hypothesis = 0.05  # 5% significance level

# Theoretical mean weekly cost (null hypothesis)
# W = 1000 + 5X, with X = 600
theoretical_mean = 1000 + 5 * mean_X
print(f"\nTheoretical mean weekly cost (μ₀): Rs. {theoretical_mean}")

# Standard deviation of weekly cost
# W = 1000 + 5X, so σ_W = 5 × σ_X
population_std_cost = 5 * population_std_X
print(f"Population standard deviation of cost (σ): Rs. {population_std_cost}")

print(f"Sample mean weekly cost (x̄): Rs. {sample_mean_cost}")
print(f"Sample size (n): {sample_size}")
print(f"Significance level (α): {alpha_hypothesis}")
```

```python
# Step 1: State the hypotheses
```

```python
print("H₀ (Null Hypothesis): μ = Rs. 4,000 (Cost model is accurate)")
print("H₁ (Alternative Hypothesis): μ > Rs. 4,000 (Costs are higher than model)")
print("This is a RIGHT-TAILED test because owners claim costs are HIGHER")
```

```python
# Step 2: Calculate the test statistic
# z = (x̄ - μ) / (σ / √n)
z_statistic = (sample_mean_cost - theoretical_mean) / (population_std_cost / math.sqrt(sample_size))
print(f"z = (x̄ - μ) / (σ / √n)")
print(f"z = ({sample_mean_cost} - {theoretical_mean}) / ({population_std_cost} / √{sample_size})")
print(f"z = ({sample_mean_cost - theoretical_mean}) / ({population_std_cost} / {math.sqrt(sample_size):.2f})")
print(f"z = {sample_mean_cost - theoretical_mean} / {population_std_cost / math.sqrt(sample_size):.2f}")
print(f"z = {z_statistic:.4f}")
```

```python
# Step 3: Determine critical value and p-value
```

```python
# Critical value for right-tailed test at α = 0.05
z_critical_hypothesis = stats.norm.ppf(1 - alpha_hypothesis)
print(f"Critical z-value (z_α) at α = 0.05: {z_critical_hypothesis:.4f}")
```

```python
# p-value (right-tailed)
p_value = 1 - stats.norm.cdf(z_statistic)
print(f"P-value: {p_value:.6f}")
```

```python
# Step 4: Make a decision
```

```python
print(f"Decision Rule (Critical Value): Reject H₀ if z > {z_critical_hypothesis:.4f}")
print(f"Decision Rule (P-value): Reject H₀ if p-value < {alpha_hypothesis}")

print(f"\nComparison:")
print(f"Test statistic z = {z_statistic:.4f} vs Critical value = {z_critical_hypothesis:.4f}")
print(f"P-value = {p_value:.6f} vs α = {alpha_hypothesis}")

if z_statistic > z_critical_hypothesis:
    print("\n✓ z > z_critical → REJECT the null hypothesis")
else:
    print("\n✗ z < z_critical → FAIL TO REJECT the null hypothesis")

if p_value < alpha_hypothesis:
    print("✓ p-value < α → REJECT the null hypothesis")
else:
    print("✗ p-value > α → FAIL TO REJECT the null hypothesis")
```

```python
# Step 5: Conclusion
```

```python
if z_statistic > z_critical_hypothesis:
    print("""CONCLUSION: There is sufficient evidence at the 5% significance level
to support the restaurant owners' claim that weekly operating costs 
are higher than the theoretical model suggests.

Since the sample mean (Rs. 3,050) is actually LOWER than the theoretical 
mean (Rs. 4,000), the test statistic is negative. This indicates that 
the observed costs are NOT higher, but lower. The owners' claim is 
not supported by the data.""")
else:
    print("""CONCLUSION: There is NOT sufficient evidence at the 5% significance level
to support the restaurant owners' claim that weekly operating costs 
are higher than the theoretical model suggests.

The test statistic (z = -3.80) is negative, meaning the sample mean 
(Rs. 3,050) is actually LOWER than the theoretical mean (Rs. 4,000).
This suggests that current operating costs are below the model's 
prediction, contradicting the owners' claim of higher costs.""")
```
