```python
import numpy as np
import pandas as pd  
import random
```

```python
# exercise1 -- Prime number


def is_prime(n):

    # Prime numbers must be greater than 1
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    
    return True

# Test the function with some examples
test_numbers = [1, 2, 3, 4, 5, 10, 17, 20, 23, 100]

print("Prime Number Checker")
print("-" * 40)
for num in test_numbers:
    if is_prime(num):
        print(f"{num} is a prime number")
    else:
        print(f"{num} is not a prime number")
```

```python
# Exercise 2 --- Product of random Number

# Generate two random integers between 1 and 10
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

# Ask the user for the product
print(f"What is {num1} * {num2}?")
user_answer = int(input("Your answer: "))

# Check if the answer is correct
correct_answer = num1 * num2

if user_answer == correct_answer:
    print(" Correct! Great job!")
else:
    print(f" Incorrect. The correct answer is {correct_answer}.")
```

```python
# Program to print the squares of all even numbers between 100 and 200
for num in range(100, 201):  # range is up to 201 to include 200
    if num % 2 == 0:  # check if the number is even
        square = num ** 2
        print(f"The square of {num} is {square}")
```

```python
# Exercise--4---Hospital Billing System
days, tests, medicine = 4, 3, 2400

room = days * 2000
doctor = 1500
lab = tests * 300
total = room + doctor + lab + medicine

if total > 10000:
    discount = total * 0.10
    final = total - discount
else:
    discount = 0
    final = total

print(f"Total Bill: ₹{total:,.2f}")
print(f"Discount: ₹{discount:,.2f}")
print(f"Final Amount: ₹{final:,.2f}")
```
