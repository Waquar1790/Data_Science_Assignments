```python
# 6 Grade based on Marks

marks = int(input("Enter marks: "))
if marks >= 90:
    grade = 'A'
elif marks >= 75:
    grade = 'B'
elif marks >= 50:
    grade = 'C'
else:
    grade = 'Fail'
print("Grade:", grade)
```

```python
# 7 check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

num = int(input("Enter a number: "))
if num > 0:
    print("Positive number")
    if num % 2 == 0:
        print("It is also even")
    else:
        print("It is odd")
elif num == 0:
    print("Zero")
else:
    print("Negative number")
```

```python
# Section C: Exploring Pandas

# Create DataFrame with 10 rows
data = {
    'Name': ['Krish', 'Ram', 'Sita', 'David', 'Sumit', 
             'Aman', 'dev', 'Anish', 'Raghav', 'Waquar'],
    'Age': [25, 30, 35, 40, 28, 33, 29, 45, 31, 27],
    'City': ['Delhi', 'Chandighar', 'Mumbai', 'Ahmad Nagar', 'Pune',
             'Patna', 'jaipur', 'hydrabad', 'Nasik', 'Surat'],
    'Salary': [50000, 60000, 70000, None, 55000, 65000, None, 80000, 62000, 58000]
}
```

```python
df = pd.DataFrame(data)
```

```python
# 1. Info and describe
print("Information")
df.info()
print("\n describe")
print(df.describe())
```

```python
# 2. Select Name and City columns

print("\ Selected columns (Name, City)")

print(df[['Name', 'City']])
```

```python
# 3. Drop City column
df_no_city = df.drop('City', axis=1)
print("\n After dropping City column")
print(df_no_city.head())
```

```python
# 4. Fill missing Salary with mean

mean_salary = df['Salary'].mean()
df['Salary'].fillna(mean_salary, inplace=True)
print("\n After filling missing Salary values")
print(df)
```

```python
# 5. Remove duplicate rows (if any)
df.drop_duplicates(inplace=True)
print("\n After removing duplicates (no changes here)")
print(df)
```
